# Modelo de esfuerzo y tiempo en proyectos de software

Calculamos el esfuerzo (E) y el tiempo calendario (td) en función del tamaño de un proyecto (S), usando las siguientes fórmulas:

- Esfuerzo: `E(S) = 8 * S^0.9`
- Tiempo calendario: `td(E) = 2.4 * E^0.4`

Se grafican los valores de E para S ∈ [1, 10000] y los de td para E ∈ [1, 500].


```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_esfuerzo(s):
    return 8 * s**0.9

def calcular_tiempo_calendario(e):
    return 2.4 * e**0.4
```

```python
tamaños = np.linspace(1, 10000, 500)
esfuerzos = calcular_esfuerzo(tamaños)

plt.figure(figsize=(10, 5))
plt.plot(tamaños, esfuerzos, label="Esfuerzo (E) vs Tamaño (S)")
plt.title("Esfuerzo en función del tamaño del proyecto")
plt.xlabel("Tamaño del proyecto (S)")
plt.ylabel("Esfuerzo (E)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
esfuerzos_td = np.linspace(1, 500, 500)
tiempos_calendario = calcular_tiempo_calendario(esfuerzos_td)

plt.figure(figsize=(10, 5))
plt.plot(esfuerzos_td, tiempos_calendario, color='orange', label="Tiempo (td) vs Esfuerzo (E)")
plt.title("Tiempo calendario en función del esfuerzo")
plt.xlabel("Esfuerzo (E)")
plt.ylabel("Tiempo calendario (td)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
